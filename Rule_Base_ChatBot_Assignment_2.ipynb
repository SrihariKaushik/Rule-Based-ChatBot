```python
# Rule-Based Library Assistant
def library_assistant():
  print("--- Welcome to the College Library Assistant ---")

  while True:
      print("\n Assistant: How can I help you? (Type: 'fine', 'hours', or 'exit')")
      user_choice = input("You: ").lower()

      # Requirement 1 & 3: Rule-based logic with IF-ELSE conditions
      if user_choice == "exit":
          print("Assistant: Goodbye! Happy studying.")
          break

      elif user_choice == "hours":
          print("Assistant: We are open 9:00 AM to 5:00 PM daily.")

      elif user_choice == "fine":
          # Requirement 4: Exception Handling
          try:
              days = int(input("Assistant: How many days late is the book? "))

              # Requirement 2: Mathematical Calculation
              # Calculation: $5 flat fee + $2 per day
              total_fine = 5 + (days * 2)

              if days <= 0:
                  print("Assistant: No fine! You returned it on time.")
              else:
                  print(f"Assistant: Your total fine is ${total_fine}.")

          except ValueError:
              print("Assistant: Error! Please enter a number only (e.g., 5).")

      else:
          print("Assistant: I didn't understand. Please type 'fine' or 'hours'.")


library_assistant()
```
